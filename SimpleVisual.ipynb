```python
from google.colab import drive
drive.mount('/content/drive')

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# List of image paths
image_paths = [
    '/content/real_spectrograms/linus-original.png',
    '/content/real_spectrograms/biden-original.png',
    '/content/real_spectrograms/margot-original.png'
]

# Display the images
plt.figure(figsize=(12, 6 * len(image_paths)))

for i, image_path in enumerate(image_paths):
    img = mpimg.imread(image_path)
    plt.subplot(len(image_paths), 1, i + 1)
    plt.imshow(img, aspect='auto')
    plt.axis('off')
    plt.title(f'NaturalSpectogram {i+1}')

plt.show()
```
